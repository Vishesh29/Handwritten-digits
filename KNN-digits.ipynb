```python
from sklearn.datasets import load_digits   #Sklearn Inbuilt Digits Dataset
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier    
```

```python
dataset=load_digits()

data=dataset.data
images=dataset.images
target=dataset.target
print('data shape:',data.shape)
print('images shape:',images.shape)
print('target shape:',target.shape)
```

```python
plt.imshow(images[0],cmap='gray')
```

```python
plt.imshow(data[0].reshape(1,64),cmap='gray')
```

```python
print(images[1016])
print(data[1016])
print(target[1016])
```

```python
train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2,random_state=42)
```

```python
#import KNN Algorithm with neighors=3
model=KNeighborsClassifier(n_neighbors=3)

model.fit(train_data,train_target)       #training the KNN Algorithm
```

```python
print('train data size:',train_data.shape)
print('test data size:',test_data.shape)
```

```python
#testing the KNN Algorithm
predicted_target=model.predict(test_data)
```

# Evaluating the Algorithm

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(test_target,predicted_target)
print('Accuracy:',acc)
```
